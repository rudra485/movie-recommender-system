```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
def convert3(cast_list):
    # Check if the input is a list
    if isinstance(cast_list, list):
        # Process each cast member in the list
        # This is a placeholder for whatever processing you need to do
        # For example, return the first 3 cast members if available
        return cast_list[:3] if len(cast_list) >= 3 else cast_list
    else:
        # Handle the case where cast_list is not a list
        # For example, return an empty list or the input as is
        return []

# Apply the fixed function to the cast column
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

## Stemming

Before vectorising, we reduce every word in `tags` to its root form using
**Porter Stemmer**. This means *'loving'*, *'loved'*, *'loves'* all become *'love'*,
so movies that share the same concept (but use different inflections) are
correctly recognised as similar.

```python
import nltk
from nltk.stem.porter import PorterStemmer

nltk.download('punkt', quiet=True)   # only needed once
ps = PorterStemmer()
```

```python
def stem(text):
    """
    Reduce each word to its root form with Porter Stemmer.
    e.g. 'actions' -> 'action', 'fighting' -> 'fight'
    """
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
# Apply stemming to the tags column
new['tags'] = new['tags'].apply(stem)
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

## TF-IDF Vectorizer

**Why add TF-IDF?**

| | `CountVectorizer` | `TfidfVectorizer` |
|---|---|---|
| Weighting | Raw count | Count × Inverse Document Frequency |
| Common words (e.g. *love*, *world*) | Over-represented | Automatically down-weighted |
| Rare, distinctive words | Under-valued | Boosted (high IDF score) |
| Best for | Short keyword bags | Rich natural-language tags |

TF-IDF rewards words that are *specific* to a small number of films,
giving a sharper cosine-similarity separation. Both vectorizers are kept
below so you can compare results directly.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# sublinear_tf=True applies log(1+tf) dampening — a word appearing 10×
# is not treated as 10× more important than one appearing once.
tfidf = TfidfVectorizer(max_features=5000, stop_words='english', sublinear_tf=True)
```

```python
tfidf_vector = tfidf.fit_transform(new['tags']).toarray()
```

```python
tfidf_vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
tfidf_similarity = cosine_similarity(tfidf_vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Batman')
```

```python
def recommend_tfidf(movie):
    """
    Recommend using the TF-IDF similarity matrix.
    Rare, descriptive words in the tags carry more weight,
    so recommendations reflect thematic nuance better than
    the plain count-based approach.
    """
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(tfidf_similarity[index])),
                        reverse=True, key=lambda x: x[1])
    print(f"TF-IDF recommendations for '{movie}':")
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend_tfidf('Batman')
```
